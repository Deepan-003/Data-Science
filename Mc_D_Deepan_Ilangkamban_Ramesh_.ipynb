### Outlier Detection ###

```python
outlier =np.arange(1,len(mc_d_num.columns))
outlier
plt.figure(figsize=(30,30))
for i in outlier:
    plt.subplot(5,4,i)
    plt.boxplot(mc_d_num.iloc[:,(i-1)])
    plt.title(mc_d_num.columns[i-1], fontsize=20)
```

### The key take away from the above box plots are : ###

almost every variables has an outlier 

### Variables with outliers : ###

Calories, Calories from fat, Total Fat, Total Fat (% Daily Value), Trans Fat, Cholesterol, Cholesterol (% Daily Value), Sodium, Sodium (% Daily Value), Carbohydrates, Carbohydrates (% Daily Values), Dietary Fiber (% Daily Value), Sugars, Protein, Vitamin A (% Daily Value), Vitamin C (% Daily Value), Calcium (% Daily Values) 

### Variables Without Outliers : ###

Saturated Fat, Saturated Fat (% Daily Value), Dietary Fiber, Dietery Fiber (% Daily Value) 

### Observation ###

Variables with outliers such as vitamin A, C, Calories and Cholesterol, Trans Fat are needed in some form for human health benifits.

Variables without outliers such as Saturated fats produced by animals fats and Dietery Fiber produced by some veggies. So most of the people are scattered in these classes

### Correlation ###

```python
correlation = mc_d_num.corr()
correlation.head()
```

```python
sns.heatmap(correlation, cmap='plasma')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(correlation[(correlation >= 0.90) | (correlation <= -0.90)], 
            cmap='plasma', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8})
```

### Key Take aways ###

***The variables having high values of correlations are***

1) Dietery Fiber (% Daily Value) and Dietery Fiber 

2) Calories from Fat and Calories

3) Total Fat and Calories

4) Total Fat (% Daily Values) and Calories 

***where every variables having highest correlation is entitiled with calories from these customers are so selective in the products which they are purchasing for health benifits***

***So, these are highly correlated with calories***
